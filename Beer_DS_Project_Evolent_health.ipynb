```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import re
```

```python
tokenizer=AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model=AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
tokens=tokenizer.encode("I like icecream, i love tea",return_tensors='pt')
tokens
```

```python
tokenizer.decode(tokens[0])
```

```python
result=model(tokens)
```

```python
result
```

```python
int(torch.argmax(result.logits))+1
```

```python
result.logits
```

```python
beer['review_text']
```

```python
beer.isnull().sum()
```

```python
beer_clean_text=beer[beer['review_text'].notnull()]
beer_clean_text.head()
beer_clean_text.shape
```

```python
beer_clean_text['review_text'][0]
```

```python
def sentiment_score(review):
    tokens=tokenizer.encode(review,return_tensors='pt')
    result=model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(beer_clean_text['review_text'][0])
```

```python
beer_clean_text.shape

beer_clean_text.reset_index(inplace=True,drop=True)
beer_clean_text.head()
```

```python
beer_clean_text2=beer_clean_text.sample(10000)
beer_clean_text2.head()
```

```python
beer_clean_text1['text_sentiment_score']=beer_clean_text1['review_text'].apply(lambda x: sentiment_score(x[:512]))
```

```python
beer_clean_text1
```

```python
beer_clean_text2['text_sentiment_score']=beer_clean_text2['review_text'].apply(lambda x: sentiment_score(x[:512]))
```

```python
beer_clean_text3=beer_clean_text2.copy()
```
